```python
def evaluate(solution, goal):
    score = 0
    for i in range(len(solution)):
        score += abs(ord(solution[i]) - ord(goal[i]))
    return score
```

```python
def mutation(solution):
    solution_ = [i for i in solution]
    solution_[random.choice(range(0,len(solution)))] = random.choice(string.printable)
    solution = ''.join(solution_)
    return solution
```

### Hill Climbing & Greedy Best Search

Structure:
1. maze
2. use tuple (x,y) as location
3. function to check possible path
4. closed list as travelled path. open list as choices
5. evaluate (x,y) location in open list to find min distance, return iloc of open list
6. if goal not reach, and is deadend, will go back to open list and pick min again

```python
class aStar():
    def __init__(self):
        self.start = (0,0)
        self.end = None
        self.self_loc = None
        self.wall_list = []
        self.open_list = []
        self.closed_list = []
        self.grid = None
        
    def maze(self, row,col,wall_no):
        """Always (0,0) as starting point and (row,col) as ending point. Wall is random."""
        grid = []
        for i in range(row):
            grid.append([0 for j in range(col)])
        self.end = (row-1, col-1)
        
        #randomize wall
        for i in range(wall_no):
            wall = (row_, col_) = random.choice(range(row)), random.choice(range(col))
            while wall in self.wall_list or wall == self.start or wall == self.end:
                wall = (row_, col_) = random.choice(range(row)), random.choice(range(col))
            self.wall_list.append(wall)
            grid[row_][col_] = 1
        
        self.grid = grid
        return grid
    
    
```
